```python
lst = [3, 5, 7, 13]
all_prime = True
for num in lst:
    if num <= 1:
        all_prime = False
        break
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            all_prime = False
            break
print(all_prime)
```
