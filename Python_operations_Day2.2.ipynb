```python
classes_held = int(input("Enter number of classes held: "))
classes_attended = int(input("Enter number of classes attended: "))

attendance = (classes_attended / float(classes_held)) * 100
print(f"Attendance: {attendance:.2f}%")

if attendance >= 75:
    print("Allowed to sit in exam.")
else:
    print("Not allowed to sit in exam.")
```

```python
classes_held = int(input("Enter number of classes held: "))
classes_attended = int(input("Enter number of classes attended: "))
medical_cause = input("Do you have a medical cause? (Y/N): ")

attendance = (classes_attended / float(classes_held)) * 100
print(f"Attendance: {attendance:.2f}%")

if attendance >= 75:
    print("Allowed to sit in exam.")
else:
    if medical_cause.upper() == 'Y':
        print("Allowed to sit in exam due to medical cause.")
    else:
        print("Not allowed to sit in exam.")
```

```python
n = int(input("Enter a number: "))
rev = 0
while n > 0:
    rev = rev * 10 + n % 10
    n //= 10
print("Reversed number:", rev)
```

```python
even_sum = 0
odd_sum = 0
for _ in range(5):  # Change range for more inputs
    num = int(input("Enter an integer: "))
    if num % 2 == 0:
        even_sum += num
    else:
        odd_sum += num
print("Even sum:", even_sum)
print("Odd sum:", odd_sum)
```

```python
n = int(input("Enter a number: "))
if n > 1:
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            print("Not a prime number.")
            break
    else:
        print("Prime number.")
else:
    print("Not a prime number.")
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b:
    a, b = b, a % b
print("HCF is:", a)
```

```python
pos = neg = zero = 0
while True:
    num = int(input("Enter a number (or -999 to stop): "))
    if num == -999:
        break
    if num > 0:
        pos += 1
    elif num < 0:
        neg += 1
    else:
        zero += 1
print("Positive:", pos, "Negative:", neg, "Zero:", zero)
```

```python
nums = []
while True:
    num = int(input("Enter a number (or -999 to stop): "))
    if num == -999:
        break
    nums.append(num)

print("Largest:", max(nums))
print("Smallest:", min(nums))
```

```python
for num in range(1, 501):
    sum = 0
    temp = num
    while temp > 0:
        digit = temp % 10
        sum += digit ** 3
        temp //= 10
    if num == sum:
        print(num)
```

```python
import random
number = random.randint(1, 100)

while True:
    guess = int(input("Guess the number: "))
    if guess < number:
        print("Too low, try again.")
    elif guess > number:
        print("Too high, try again.")
    else:
        print("Correct guess!")
        break
```
